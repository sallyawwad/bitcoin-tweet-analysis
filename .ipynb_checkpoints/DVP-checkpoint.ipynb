### Text Preprocessing and Graphical Annotation Functions

```python
def offset_png(x, y, path, ax, zoom, offset):
    '''For adding other .png images to the graph.
    source: https://stackoverflow.com/questions/61971090/how-can-i-add-images-to-bars-in-axes-matplotlib'''
    #add a PNG images to a matplotlib graph (creates an OffsetImage object, and adds it to the specified axis at the given coordinates.) 
    img = plt.imread(path)
    im = OffsetImage(img, zoom=zoom)
    im.image.axes = ax
    x_offset = offset
    ab = AnnotationBbox(im, (x, y), xybox=(x_offset, 0), frameon=False,
                        xycoords='data', boxcoords="offset points", pad=0)
    ax.add_artist(ab)
    

#This function adds text labels to a seaborn barplot
def show_values_on_bars(axs, h_v="v", space=0.4):
    '''Plots the value at the end of the a seaborn barplot.
    axs: the ax of the plot
    h_v: weather or not the barplot is vertical/ horizontal'''
    
    def _show_on_single_plot(ax):
        if h_v == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height()
                value = int(p.get_height())
                ax.text(_x, _y, format(value, ','), ha="center") 
        elif h_v == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height()
                value = int(p.get_width())
                ax.text(_x, _y, format(value, ','), ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _show_on_single_plot(ax)
    else:
        _show_on_single_plot(axs)
        


def emoji_extractor(string, remove=False):
#removed from the text (True) or just extracted (False).
#re.compile: compiles this pattern into a regular expression object which represents Unicode ranges representing various categories of emojis.
    '''Removes Emoji from a text.'''
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    if remove == False:
        # Extract emoji
        return emoji_pattern.findall(string)
    else:
        # Remove emoji from text
        return emoji_pattern.sub(r'', string)

def clean_emoji(x):
    if len(x) == 0:
        return ''
    else:
        return x[0]
    
    
    
def clean_tweets(df):
    '''Returns the dataframe with the tweet column cleaned.'''
    
    # ----- Remove \n, \t, \xa0 -----
    df['tweet'] = df['tweet'].apply(lambda x: x.replace('\n', ''))
    df['tweet'] = df['tweet'].apply(lambda x: x.replace('\xa0', ''))
    df['tweet'] = df['tweet'].apply(lambda x: x.replace('\t', ''))
    
    # ----- Remove pic.twitter and http:// + https:// links -----
    df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'http\S+', '', x))
    df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'https\S+', '', x))
    df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'pic.twitter\S+', '', x))
    
    # ----- Remove mentions and hashtags -----
    df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'#\S+', '', x))
    df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'@\S+', '', x))
    
    # ----- Extract Emojis and Remove from Tweet -----
    df['tweet_emojis'] = df['tweet'].apply(lambda x: emoji_extractor(x, remove=False))
    df['tweet_emojis'].replace('', np.nan, inplace=True)
#     df["tweet_emojis"] = df["tweet_emojis"].apply(lambda x: clean_emoji(x))
    
    df['tweet'] = df['tweet'].apply(lambda x: emoji_extractor(x, remove=True))
    
    # ----- Strip of whitespaces -----
    df['tweet'] = df['tweet'].apply(lambda x: x.strip())
    df['tweet'] = df['tweet'].apply(lambda x: ' '.join(x.split()))
    
    # ----- Remove punctuation & Make lowercase -----
    df['tweet'] = df['tweet'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
    df['tweet'] = df['tweet'].apply(lambda x: x.lower())
    
    return df
```

### Loading datasets and merging them based on date

```python
# Load Tweets dataset
tweets = pd.read_csv("C:\\Users\\lenovo\\Desktop\\DVProject\\TweetsElonMusk.csv")
tweets = tweets[["id", "date", "time", "username", 
                 "tweet", "mentions", "urls", "photos", "replies_count", 
                 "retweets_count", "likes_count", "hashtags", "link"]]

# Create new features
tweets["year"] = tweets["date"].apply(lambda x: x.split("-")[0])

# Clean Tweets
tweets = clean_tweets(df=tweets)

# Load Bitcoin dataset
bitcoin = pd.read_csv("C:\\Users\\lenovo\\Desktop\\DVProject\\btc.csv")

# Convert date columns to datetime
tweets['date'] = pd.to_datetime(tweets['date'])
bitcoin['date'] = pd.to_datetime(bitcoin['date'])

# Determine common date range
common_start_date = max(tweets['date'].min(), bitcoin['date'].min())
common_end_date = min(tweets['date'].max(), bitcoin['date'].max())

# Trim datasets to common date range
tweets_trimmed = tweets[(tweets['date'] >= common_start_date) & (tweets['date'] <= common_end_date)]
bitcoin_trimmed = bitcoin[(bitcoin['date'] >= common_start_date) & (bitcoin['date'] <= common_end_date)]

# Merge datasets based on date
merged_df = pd.merge(tweets_trimmed, bitcoin_trimmed, on='date', how='inner')
#sort dates
tweets_sorted = tweets.sort_values(by='date')
bitcoin_sorted = bitcoin.sort_values(by='date')

merged_df = pd.merge(tweets_sorted, bitcoin_sorted, on='date', how='inner')

```

### Break down of the datasets

#### I. Bitcoin dataset :
Here is a breakdown of the columns in this dataset (Jul 2010-Mar 2024):
1. **Date**: This column represents the date when the trading activity occurred.

2. **Open**: This column represents the opening price of Bitcoin for the given date.

3. **High**: This column represents the highest price of Bitcoin reached during the trading session on the given date.

4. **Low**: This column represents the lowest price of Bitcoin reached during the trading session on the given date.

5. **Close**: This column represents the closing price of Bitcoin for the given date. The closing price is the price at which the last trade occurred during a trading session.

6. **Volume**: This column represents the total amount of Bitcoin traded (bought or sold) during the trading session on the given date. It typically indicates the level of trading activity for that day.

#### II. Elon Musk's tweets dataset:
Here is a breakdown of the columns in this dataset (2010 - 2021):

1. **id**: Unique identifier for each tweet.
2. **date**: Date when the tweet was posted.
3. **time**: Time when the tweet was posted.
4. **username**: Username of the user who posted the tweet.
5. **tweet**: The actual text content of the tweet.
6. **mentions**: Usernames mentioned in the tweet.
7. **urls**: URLs included in the tweet.
8. **photos**: Photos attached to the tweet.
9. **replies_count**: Number of replies to the tweet.
10. **retweets_count**: Number of retweets of the tweet.
11. **likes_count**: Number of likes (favorites) for the tweet.
12. **hashtags**: Hashtags included in the tweet.
13. **link**: Link to the tweet.
14. **year**: Year when the tweet was posted (extracted from the date).
15. **tweet_emojis**: Emojis extracted from the tweet.ge of the translation, if applicable.

### Popularity 📈


```python
# Get Popularity Information
popularity = ["likes_count", "retweets_count", "replies_count"]
popularity_df = tweets[tweets["year"]!="2021"].groupby("year").agg({popularity[0] : 'sum',
                                                                    popularity[1] : 'sum',
                                                                    popularity[2] : 'sum',
                                                                    'tweet' : 'count'}).reset_index()
popularity_df["likes_count"] = popularity_df["likes_count"]/popularity_df["tweet"]
popularity_df["retweets_count"] = popularity_df["retweets_count"]/popularity_df["tweet"]
popularity_df["replies_count"] = popularity_df["replies_count"]/popularity_df["tweet"]

# Plot
fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(25, 15))
axs = [ax1, ax2, ax3]
plt.suptitle("Popularity", size=25)
sns.barplot(data=popularity_df, x="year", y="likes_count", lw=5, color=my_colors[0], ax=ax1)
sns.barplot(data=popularity_df, x="year", y="retweets_count", lw=5, color=my_colors[1], ax=ax2)
sns.barplot(data=popularity_df, x="year", y="replies_count", lw=5, color=my_colors[2], ax=ax3)
names = ["Average Likes", "Average Retweets", "Average Replies"]
for ax, n in zip(axs, names):
    ax.set_xlabel("", size=20)
    ax.set_ylabel(n, size=20)
    ax.get_yaxis().set_ticks([])
#     ax.title.set_text(n)
    show_values_on_bars(axs=ax, h_v="v", space=0.4)
sns.despine(left=True);
```

### Closer Look 🔍


Let's use a magnifying glass and look closer to our points in time.
We took only the last 3 tweets, then we zoomed in these data points (where they are located in x-axis and y-axis) to get more insight.

```python
# Get bitcoin info
btc_tweets = tweets[tweets["tweet"].str.contains("bitcoin")].reset_index(drop = True)
# Convert date to number
btc_tweets["date"] = btc_tweets["date"].apply(lambda x: x.timestamp())
# Get only latest bitcoin data (as Elon never tweeted before 2018)
btc_prices = bitcoin.sort_values("date", ascending=False).head(2000000)
timestamps = btc_tweets["date"]
for k, tweet in enumerate(btc_tweets["tweet"][:3]): print(color.BOLD + f"{k+1}." + color.END, tweet)
```

```python
# Filter tweets containing "bitcoin"
btc_tweets = tweets[tweets["tweet"].str.contains("bitcoin")].reset_index(drop=True)

# Convert date to number
btc_tweets["timestamp"] = pd.to_datetime(btc_tweets["date"]).astype('int64') // 10**9

# Sort btc_tweets DataFrame based on the 'timestamp' column
btc_tweets = btc_tweets.sort_values('timestamp', ascending=True)

# Assuming the 'bitcoin' DataFrame also has a 'date' column in datetime format
bitcoin['timestamp'] = pd.to_datetime(bitcoin['date']).astype('int64') // 10**9

# Define start and end dates
start_date = "2019-01-05"
end_date = "2021-05-29"

# Filter Bitcoin data between 2020 and 2022 and sort
btc_prices = bitcoin[(bitcoin['date'] >= start_date) & (bitcoin['date'] <= end_date)]
btc_prices = btc_prices.sort_values('date', ascending=True)

# Merge on timestamps within a certain window (e.g., +/- 1 hour)
merged = pd.merge_asof(btc_tweets, btc_prices, on='timestamp', direction='nearest', tolerance=3600)

# Filter out non-matching rows and get the first 3 entries
merged = merged.dropna(subset=['Close'])
merged = merged.head(3)  # Get only the first 3 points

# Plotting
plt.figure(figsize=(15, 8))
plt.plot(btc_prices['date'], btc_prices['Close'], label='Bitcoin Price', color=my_colors[0], linewidth=2) 
plt.scatter(merged['date_y'], merged['Close'], color='red', s=50, label='Tweet Impact', zorder=5)

# Formatting the plot
plt.title('Last 3 tweets in time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Price (USD)', fontsize=14)
plt.legend()
plt.grid(True)

# Adjust y-axis range
plt.ylim(5000, 11000)  # Set y-axis to show between $5,000 and $11,000

plt.show()

```
